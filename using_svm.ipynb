```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
cat="/content/drive/MyDrive/Dataset"
dataset=os.listdir(cat)
```

```python
lables=[i for i in range(len(dataset))]
lables_dict=dict(zip(dataset,lables))
```

```python
lables_dict
```

```python
data=[]
target=[]
thrsh=170
maxval=255
```

```python
for i in dataset:
    forder_path=os.path.join(cat,i)
    img_names=os.listdir(forder_path)
    for img_name in img_names:
        img_path=os.path.join(forder_path,img_name)
        img=plt.imread(img_path)
        im_bin=(img>thrsh)*maxval
        Image.fromarray(np.uint8(im_bin)).save('bin.RGBA')
        data.append(im_bin.flatten())
        target.append(lables_dict[i])
```

```python
data=pd.DataFrame(data)
target=pd.DataFrame(target)
```

```python
data
```

```python
target
```

```python
from sklearn.impute import SimpleImputer
ImputedModule = SimpleImputer()
ImputedX = ImputedModule.fit(data)
data = ImputedX.transform(data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.20,random_state=44, shuffle =True)
```

```python
from sklearn.svm import SVC
SVCModel = SVC(kernel = 'linear', random_state = 0)
SVCModel.fit(X_train, y_train)

print('SVCModel Train Score is : ' , SVCModel.score(X_train, y_train))
print('SVCModel Test Score is : ' , SVCModel.score(X_test, y_test))
print('----------------------------------------------------')
```

```python
from sklearn.svm import SVC
SVCModel = SVC(kernel = 'sigmoid', random_state = 0)
SVCModel.fit(X_train, y_train)
#Calculating Details
print('SVCModel Train Score is : ' , SVCModel.score(X_train, y_train))
print('SVCModel Test Score is : ' , SVCModel.score(X_test, y_test))
print('----------------------------------------------------')
```
